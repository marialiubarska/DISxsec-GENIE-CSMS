```python
def get_weight(lgE, y, coef=0.):
    w0 = weight_func.ev(lgE,y)
    w = w0*(1. + (1./w0 - 1)*(1. - coef))
    
    return w
```

```python
def interpol_lgE_y(ind_genie, ind_nugen, 
                   nbins=20,
                   title='', label='~~~',
                   color='b', alpha=1., coef=1., clim=[0.5,1.5]):
    
    cm1 = 'jet'
    
    Z = []
    N = []
    beY = []
    beE = []
    PX = []
    PY = []
    
    bin_edg_lgE = np.linspace(Emin,4.0,nbins+1)
    
    for i in range(len(bin_edg_lgE)-1):
        lgE_min = bin_edg_lgE[i]
        lgE_max = bin_edg_lgE[i+1]
        
        # select energy bin for genie
        indl = np.transpose(np.argwhere(lgE_genie[ind_genie]>lgE_min))[0]
        indr = np.transpose(np.argwhere(lgE_genie[ind_genie][indl]<=lgE_max))[0]
        ind_bin_genie = ind_genie[indl[indr]]

        # select energy bin for nugen
        indl = np.transpose(np.argwhere(lgE_nugen[ind_nugen]>lgE_min))[0]
        indr = np.transpose(np.argwhere(lgE_nugen[ind_nugen][indl]<=lgE_max))[0]
        ind_bin_nugen = ind_nugen[indl[indr]]

        # calculate weights
        w_genie = one_weight_genie[ind_bin_genie]/float(sum(one_weight_genie[ind_bin_genie]))
        w_genie_nonorm = one_weight_genie[ind_bin_genie]
        w_nugen = one_weight_nugen[ind_bin_nugen]/float(sum(one_weight_nugen[ind_bin_nugen])) 
        
        # bin genie and nugen in Y
        h_genie, bin_edg_Y = np.histogram(Y_genie[ind_bin_genie],
                                          nbins,range=[0.,1.], weights=w_genie)
        h_nugen, bin_edg_Y = np.histogram(Y_nugen[ind_bin_nugen],
                                          nbins,range=[0.,1.], weights=w_nugen)
    
        # calculate ratios and save them to array
        n_rat = h_nugen / h_genie
        N.append(n_rat)
        
        # bin genie and nugen in Y
        h_genie_nonorm, _ = np.histogram(Y_genie[ind_bin_genie],
                                          nbins,range=[0.,1.], weights=w_genie_nonorm)
        
        print '~~~~~~ ratio corr/init: ', np.sum(h_genie_nonorm*n_rat)/np.sum(h_genie_nonorm)
        
        beY.append(bin_edg_Y[:-1]+(bin_edg_Y[1]-bin_edg_Y[0])/2.)
        beE.append(np.ones(len(n_rat))*(lgE_min+(lgE_max-lgE_min)/2.))
        
    N = np.array(N)
    PX = bin_edg_lgE[:-1]
    PY = bin_edg_Y[:-1]
    PX0, PY0 = np.meshgrid(PX, PY)
    X0 = np.linspace(Emin, 4., 100)
    Y0 = np.linspace(0., 1., 100)
    Z = np.rot90(np.array(N))

    fig = plt.figure(1, figsize=(8, 6.2))
    ax = plt.gca() 
    ax.set_aspect((4.-Emin))
    ax.set_title(title, fontsize=20)
    ax.set_xlabel('lgE')
    ax.set_ylabel('Y')
    
    p = ax.imshow(Z, extent=(Emin,4,0,1), 
                     interpolation='nearest', cmap='coolwarm',
                     origin='upper', clim=clim, aspect=4.-Emin, norm=Normalize())
    
    ax.grid(color='g', linestyle='--', linewidth=0.4, alpha=0.5)
    fig.colorbar(p, label='CSMS/GENIE')#, shrink=0.85, aspect=15)
    
    save_name = 'diff_xsec_ratios_'+title.replace(' ', '_')
    plt.savefig(os.path.join(plots_save_pth,save_name+'.png'), dpi='figure')
    plt.savefig(os.path.join(plots_save_pth,save_name+'.pdf'), dpi='figure')
    
    plt.show()
    
    #!!!
    N2 = N.flatten("C")
    beE = np.array(beE).flatten("C")
    beY = np.array(beY).flatten("C")
    print len(beE), len(beY), len(N2)
    Ti = SmoothBivariateSpline(beE,beY, N2)
    
    fig = plt.figure(1, figsize=(8, 6.2))
    ax = plt.gca()
    ax.set_aspect((4.-Emin))
    ax.set_title(title+' interp. '+str(nbins)+' bins', fontsize=20)
    ax.set_xlabel('lgE')
    ax.set_ylabel('Y')
    
    p = ax.scatter(beE,beY, c=Ti.ev(beE,beY), cmap='coolwarm', s=25000./nbins,
                      norm=Normalize(vmin=clim[0], vmax=clim[1]))

    ax.grid(color='g', linestyle='--', linewidth=0.4, alpha=0.5)
    fig.colorbar(p, label='CSMS/GENIE')
    ax.legend(fontsize=16, loc=2, ncol=3)
    plt.show()
    
    return Ti
```

```python
def save_spline(wf, t):
    fo = open(os.path.join(files_save_pth,'splines_flat/'+t+'flat.pckl'),'wb')
    pickle.dump(weight_func, fo) 
    fo.close()
```
