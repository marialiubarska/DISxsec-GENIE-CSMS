```python
d_x,d_y = np.load(os.path.join(pth,'mean_y_plot_paper.npy'))

vert_err_t = np.abs(np.array(d_y['exp_err_t'])-np.array(d_y['exp']))
vert_err_b = np.abs(np.array(d_y['exp'])-np.array(d_y['exp_err_b']))
horz_err_r = np.abs(np.log10(d_x['exp_err_r'])-np.log10(d_x['exp']))
horz_err_l = np.abs(np.log10(d_x['exp'])-np.log10(d_x['exp_err_l']))

d_y_vert_err = [vert_err_b,vert_err_t]
d_x_horz_err = [horz_err_r,horz_err_l]
```

```python
def plot_means(ax, lgE, Y, w, ind,label,fmt='ko', color='b', lgE_range=[2., 4.]):
    
#     min_lgE = np.min(lgE[ind])
#     max_lgE = np.max(lgE[ind])
    
    bm, be, bn = binned_statistic(lgE[ind], Y[ind]*w[ind],
                                  bins=100, range=lgE_range, statistic='sum')
    
    bw, _, _ = binned_statistic(lgE[ind], w[ind],
                                  bins=100, range=lgE_range, statistic='sum')
    ax.plot(be[:-1]+0.5*(be[1]-be[0]),bm/bw,fmt,markersize=2, label=label, c=color, lw=3.)
    return bm/bw, be[:-1]+0.5*(be[1]-be[0])
```

```python
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(10)
gs = mpl.gridspec.GridSpec(2, 1, height_ratios=[2,1], hspace=0.) 
plt.subplot(gs[0])
ax = plt.gca()

ax.set_xlim(0.,3.)
ax.set_ylim(0.,1.0)
plt.xticks(visible=False)
ax.set_title('L5 Nu CC')

ax.set_ylabel('y')

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
bmg, be = plot_means(ax=ax, lgE=dr['lgE'], Y=dr['y'], w=dr['weights'],
                    ind=dr['CCdisNu'],label='initial',fmt='-',color='red',
                    lgE_range=[0.,3.])

bmc, be = plot_means(ax=ax, lgE=dr['lgE'], Y=dr['y'], w=dr['weights_corr'],
                    ind=dr['CCdisNu'],label='corrected',fmt='-',color='blue',
                    lgE_range=[0.,3.])

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

ax.grid(color='g', linestyle='--', linewidth=0.4, alpha=0.5)
plt.legend(loc=2, ncol=2)

plt.subplot(gs[1])
ax = plt.gca()

ax.set_xlim(0.,3.)
ax.set_ylim(0.88,1.15)
ax.set_xlabel('log10(E), GeV')
plt.ylabel('CSMS/Genie')

ax.plot(be,bmg/bmg,'red',lw=3)
ax.plot(be,bmc/bmg,'blue',lw=3)

ax.grid(color='g', linestyle='--', linewidth=0.4, alpha=0.5)


plt.savefig(os.path.join(plots_save_pth,'L5_NuCC_comp'+'.png'), dpi='figure')
plt.savefig(os.path.join(plots_save_pth,'L5_NuCC_comp'+'.pdf'), dpi='figure')
plt.show()
```
